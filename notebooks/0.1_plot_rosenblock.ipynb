## 元の勾配をとる

```python
linspace = np.linspace(-2, 2, 101)
x, y = np.meshgrid(linspace, linspace)
```

```python
z = rosenblock(x, y)
```

```python
trace = get_trace()

tr_x = trace[:, 0]
tr_y = trace[:, 1]
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.contour(x, y, z, levels=[0, 1, 5, 10, 50, 100, 1000, 2000])

plt.plot(tr_x, tr_y, '-o')

plt.scatter([1.0], [1.0], color='red', marker=(5, 1, 0), linewidths=3)
```
